```python
print("--- Smart Calculator (Type 'exit' to stop) ---")

while True:
    # 1. User se input lena
    user_input = input("\nEnter first number (or 'exit'): ").lower()

    # Check karna ke kahin user band toh nahi karna chahta
    if user_input == 'exit':
        print("Calculator band ho raha hai. Khuda Hafiz!")
        break

    try:
        # Input ko number mein badalna
        num1 = float(user_input)
        
        op = input("Enter operation (+, -, *, /): ")
        
        num2 = float(input("Enter second number: "))

        # 2. Conditionals (Calculations)
        if op == '+':
            print(f"Result: {num1 + num2}")
        elif op == '-':
            print(f"Result: {num1 - num2}")
        elif op == '*':
            print(f"Result: {num1 * num2}")
        elif op == '/':
            # 3. Logic: Divide by zero check
            if num2 == 0:
                print("Error: Cannot divide by zero!")
            else:
                print(f"Result: {num1 / num2}")
        else:
            print("Invalid Operation! Please use +, -, *, or /")

    except ValueError:
        print("Invalid input! Please enter a valid number.")
```

```python
import json

filename = 'config.json'

try:
    # 1. READ: File se purana data uthana
    with open(filename, 'r') as file:
        data = json.load(file)
        print("Purana Data:", data)

    # 2. UPDATE: Dictionary mein naya key-value pair add karna
    data['theme'] = 'dark'
    data['language'] = 'Urdu'

    # 3. WRITE/SAVE: Naye data ko wapas file mein save karna
    with open(filename, 'w') as file:
        # indent=4 se file saaf nazar aayegi
        json.dump(data, file, indent=4)
        
    print("\n Success! File update ho gayi hai.")

except FileNotFoundError:
    print(f" Error: '{filename}' nahi mili. Pehle file banayein!")
```

```python
import json
filename = "user_choice.json"
user_favclr = input("Enter your favorite color: ")
user_age = int(input("Enter your age: "))
data_to_save = {
    "favorite_color": user_favclr,
    "age": user_age
}
try:
    with open(filename, "w") as f:
        json.dump(data_to_save, f, indent=4)
    
    print(f"Mubarak ho! Data '{filename}' mein save ho gaya hai.")
except Exception as e:
    print(f" Kuch masla hua: {e}")
```
