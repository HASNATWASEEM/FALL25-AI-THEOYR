```python
import json
inventory = []

try:
    # 1. Yahan quotes lagaye: "inventory.json"
    with open("inventory.json", "r") as file:
        inventory = json.load(file)
        print("OLD DATA LOADED SUCCESFULLY")
except (FileNotFoundError, json.JSONDecodeError):
    inventory = []
    print("NEW DATA STARTING FROM HERE")

def add_item():
    # 2. 'int' hata diya taake naam likha ja sake
    item_name = input("enter item name: ") 
    quantity = input("enter quantity: ")
    
    # 3. Ye poora TRY block ab function ke ANDAR hai
    try:
        qty = int(quantity)
        new_inventory = {"item": item_name, "qty": qty}
        inventory.append(new_inventory)
        with open("inventory.json", "w") as f:
            json.dump(inventory, f, indent=4)
        print(f"{item_name} inventory mein save ho gaya!")
    except ValueError:
        print("quantity in number")

def show_inventory():
    print("\n--- Maujooda Saaman (Inventory) ---")
    if not inventory:
        print("Inventory khali hai!")
    else:
        for index, item in enumerate(inventory, 1):
            print(f"{index}. Item: {item['item']} | Qty: {item['qty']}")
    print("----------------------------------")

# 4. While loop ko function se bahar bilkul side par kar diya
while True:
    print("\nMain Menu:")
    print("1 for add new stock")
    print("2 for check inventory")
    print("3 dukan bnd kara")
    
    choice = input("Select karein (1/2/3): ")

    if choice == "1":
        add_item()
    elif choice == "2":
        show_inventory()
    elif choice == "3":
        print("Allah Hafiz! Dukan band ho gayi.")
        break
    else:
        print(" Galat option! Sahi number select karein.")
```
